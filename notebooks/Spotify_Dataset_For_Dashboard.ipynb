```python
# Load Original

spotify_original = pd.read_csv("/Users/jo186027/Downloads/spotify_playlists/spotify_dataset.csv", error_bad_lines=False)
spotify_original.columns = ['id','artist_name','track_name','playlist_name']
```

*** b.1. Get Dataframe for Unique Columns ***

```python
data = spotify_original.apply(pd.Series.nunique)
fp_unique_df = pd.DataFrame({'Columns':data.index,'Unique_Values':data.values})
fp_unique_df.to_csv('/Users/jo186027/Desktop/fp_unique_df.csv',index=False)
fp_unique_df.head()
```

*** b.2 Get Dataframe for 100 Most Popular Songs ***

```python

fp_popular_songs = spotify_original.groupby(by=['track_name'],as_index=False).agg({'id':'nunique'})
fp_popular_songs.sort_values(by='id',inplace=True,ascending=False)
fp_popular_songs.reset_index(drop=True,inplace=True)
fp_popular_songs.head()
```

```python
# save top 100
fp_songs_100 = fp_popular_songs.loc[0:100,:]
fp_songs_100.to_csv('/Users/jo186027/Desktop/fp_songs_100.csv',index=False)
```

*** b.3 Get Dataframe for 100 Most Popular Artist ***

```python

fp_popular_artists = spotify_original.groupby(by=['artist_name'],as_index=False).agg({'id':'nunique'})
fp_popular_artists.sort_values(by='id',inplace=True,ascending=False)
fp_popular_artists.reset_index(drop=True,inplace=True)
fp_popular_artists.head(10)
```

```python
# save top 100 artist
fp_artists_100 = fp_popular_artists.loc[0:100,:]
fp_artists_100.to_csv('/Users/jo186027/Desktop/fp_artists_100.csv',index=False)
```

*** b.4 Get Dataframe for Top 1000 Artist with Country of Origin ***

```python
# Scrape Country of Origin or Birthplace of Artist in Wikipedia
top_artist_country = pd.DataFrame(columns=['artist_name','id','country'])

i1 = 0
i2 = 50

for x in range(20):
    xx = fp_popular_artists.loc[i1:i2,:]
    xx['country'] = xx['artist_name'].apply(country_scraper_wrapper)
    top_artist_country = top_artist_country.append(xx)
    i1 += 50
    i2 += 50
```

```python
#
top_artist_country.fillna(value="None", inplace=True)
top_artist_country
```

```python
top_artist_country.country.value_counts()
```

```python
# Aggregate same countries that was abbreviated
us_list = ['U.S.','U.S','US','Washington, U.S.','New York City, U.S.','UnitedStates','Michigan, U.S.']
uk_list = ['England, UK','ENgland','Surrey, England','UK','U.K.','United Kingdom']
au_list = ['Victoria, Australia']
sc_list = ['East Ayrshire, Scotland']
ir_list = ['Dublin, Ireland','Northern Ireland']

for i in top_artist_country.index:
    if top_artist_country.loc[i,'country'] in us_list:
        top_artist_country.loc[i,'country'] = 'United States'
    elif top_artist_country.loc[i,'country'] in uk_list:
        top_artist_country.loc[i,'country'] = 'England'
    elif top_artist_country.loc[i,'country'] in au_list:
        top_artist_country.loc[i,'country'] = 'Australia'
    elif top_artist_country.loc[i,'country'] in sc_list:
        top_artist_country.loc[i,'country'] = 'Scotland'
    elif top_artist_country.loc[i,'country'] in ir_list:
        top_artist_country.loc[i,'country'] = 'Ireland'
```
